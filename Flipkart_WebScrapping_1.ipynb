```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://www.flipkart.com/search?q=phones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.brand%255B%255D%3DAPPLE"
```

```python
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36','Accept-Language':'en-US,en;q=0.5'})
```

```python
# HTTP request

webpage = requests.get(URL,headers=HEADERS)
```

```python
# Getting response code as 200 is good to go()
webpage
```

```python
# This will give all the content in bytes format and we have to convert it into html format
type(webpage.content)
```

```python
# This will convert into proper html format

soup = BeautifulSoup(webpage.content,"html.parser")
```

```python
# Fetch links as type of link objects

links = soup.find_all("a",attrs={'class':'_1fQZEK'})
```

```python
links
```

```python
link = links[0].get('href')
```

```python
product_list = "https://www.flipkart.com" + link
```

```python
product_list
```

```python
new_webpage = requests.get(product_list,headers=HEADERS)
```

```python
new_webpage

# Request was sucessful
```

```python
# Soup object containing all the data

new_soup = BeautifulSoup(new_webpage.content,"html.parser")
```

```python
new_soup
```

```python
# Title or the name of the product

new_soup.find("span",attrs={'class':'B_NuCI'}).text
```

```python
# Price of the product

new_soup.find("div",attrs={'class':'_30jeq3 _16Jk6d'}).text
```

```python
# Rating of the product out of 5

new_soup.find("div",attrs={'class':'_3LWZlK'}).text
```
